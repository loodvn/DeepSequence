```python
# Looking at DMS mapping and comparing to original DeepSeq MSAs
```

```python
import os

import pandas as pd
```

```python
dms_mapping = pd.read_csv("DMS_mapping_20220109.csv")
dms_mapping.head()
```

```python
file_list = os.listdir("pascal_deepseq_alignments_dir/")
file_list = sorted(file_list)
file_list
```

```python
dms_mapping['in_deepseq_MSA'] = dms_mapping['UniProt_ID'].apply(lambda s: any([m.startswith(s) for m in file_list]))
dms_mapping.head()
```

### Export DeepSeq benchmark subset:

```python
deepseq_benchmark = dms_mapping[dms_mapping['in_deepseq_benchmark'].isin(["Yes"])]
deepseq_benchmark.head()
```

```python
deepseq_benchmark = dms_mapping[dms_mapping['in_deepseq_benchmark'].isin(["Yes"])].reset_index().rename(columns={'index': 'original_index'})
deepseq_benchmark.to_csv("DMS_mapping_deepseq_20220109.csv")
print(len(deepseq_benchmark), "rows")
deepseq_benchmark.head()
```

## Count number of matches

```python
print(f"Number of MSA matches: {dms_mapping['in_deepseq_MSA'].sum()} out of {len(dms_mapping)}")
print("Mismatches in total benchmark:")
dms_mapping[~dms_mapping['in_deepseq_MSA']]
```

```python
dms_deepseq_subset = deepseq_benchmark[["original_index","DMS_id", "DMS_filename", "MSA_filename", "DMS_phenotype_name", "UniProt_ID", "in_deepseq_benchmark", 'in_deepseq_MSA']]
dms_deepseq_subset
```

```python
print("MSAs not found in deepseq mappings:", len(dms_deepseq_subset[~dms_deepseq_subset['in_deepseq_MSA']]))
print("original indices: ", dms_deepseq_subset[~dms_deepseq_subset['in_deepseq_MSA']]['original_index'].values)
print("New indices:", dms_deepseq_subset[~dms_deepseq_subset['in_deepseq_MSA']].index.values)
dms_deepseq_subset[~dms_deepseq_subset['in_deepseq_MSA']]
```
